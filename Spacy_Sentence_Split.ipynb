```python
#Splitting sentences using Spacy
import spacy
```

```python
#Let us load the smaller english language package of Spacy
lingua = spacy.load('en_core_web_sm')
```

```python
with open('Large_Text.text') as f:
    text = f.read()
```

```python
print(text[:5000])
```

```python
doc = lingua(text[:5000])
```

```python
for sentences in doc.sents:
    print(sentences)
```
